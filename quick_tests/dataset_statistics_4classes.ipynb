```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import matplotlib.colors as mcolor
import scipy.stats as stats
from scipy.interpolate import make_interp_spline
```

```python
## for robosense

train_data = '/home/cws-ml-lab/mmdetection3d_for_rail/data/robosense_cls/gt_infos_database_train.pkl'
val_data = '/home/cws-ml-lab/mmdetection3d_for_rail/data/robosense_cls/gt_infos_database_val.pkl'
train_data = pickle.load(open(train_data, 'rb'))
val_data = pickle.load(open(val_data, 'rb'))

data_list = train_data['data_list'] + val_data['data_list']

bboxes_ped = []
bboxes_cyc = []
bboxes_rv = []
bboxes_train = []
for object in data_list:
    if object['name'] == 'Pedestrian':
        bboxes_ped.append(object['box3d_lidar'])
    elif object['name'] == 'Cyclist':
        bboxes_cyc.append(object['box3d_lidar'])
    elif object['name'] == 'RoadVehicle':
        bboxes_rv.append(object['box3d_lidar'])
    elif object['name'] == 'Train':
        bboxes_train.append(object['box3d_lidar'])


# per object type median
bboxes_ped = np.array(bboxes_ped)
print('pedestrian median:', np.median(bboxes_ped, axis=0))
ped_median = np.median(bboxes_ped, axis=0)
ped_median_size = ped_median[3:6]

bboxes_cyc = np.array(bboxes_cyc)
print('cyclist median:', np.median(bboxes_cyc, axis=0))
cyc_median = np.median(bboxes_cyc, axis=0)
cyc_median_size = cyc_median[3:6]

bboxes_rv = np.array(bboxes_rv)
print('road vehicle median:', np.median(bboxes_rv, axis=0))
rv_median = np.median(bboxes_rv, axis=0)
rv_median_size = rv_median[3:6]

bboxes_train = np.array(bboxes_train)
print('train median:', np.median(bboxes_train, axis=0))
train_median = np.median(bboxes_train, axis=0)
train_median_size = train_median[3:6]

print([round(i, 2) for i in ped_median_size])
print([round(i, 2) for i in cyc_median_size])
print([round(i, 2) for i in rv_median_size])
print([round(i, 2) for i in train_median_size])

print([[round(i, 2) for i in ped_median_size], [round(i, 2) for i in cyc_median_size], [round(i, 2) for i in rv_median_size], [round(i, 2) for i in train_median_size]])
```
